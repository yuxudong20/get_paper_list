```python
import openreview

# API V2
client = openreview.api.OpenReviewClient(
    baseurl='https://api2.openreview.net',
    username='XXXX', # openreview的username
    password='XXXX'  # openreview的password
)
# submissions = client.get_all_notes(content={'venueid':'ICML.cc/2023/Conference'} )
submissions = client.get_all_notes(content={'venueid':'ICLR.cc/2023/Conference'} )
# submissions = client.get_all_notes(content={'venueid':'robot-learning.org/CoRL/2024/Conference'})
```

```python
papers = []
for accepted_submission in submissions:
    print(accepted_submission.number, accepted_submission.content['title']['value'])
    title = accepted_submission.content['title']['value']
    authors = accepted_submission.content['authors']['value']
    # keywords = accepted_submission.content['keywords']['value']
    abstract = accepted_submission.content['abstract']['value']
    nurl = f"https://openreview.net/forum?id={accepted_submission.id}"

    paper = {
                'title': title,
                'authors': authors,
                'year': accepted_submission.content['venue']['value'],
                'abstract': abstract,
                # 'keywords': keywords,
                'url': nurl,
            }
    papers.append(paper)
```

```python
# 生成HTML文件
paper_dir = submissions[0].content['venue']['value']
with open(f'{paper_dir}.html', 'w', encoding='utf-8') as htmlfile:
    htmlfile.write('''
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Papers</title>
        <style>
            table {
                width: 100%;
                border-collapse: collapse;
            }
            th, td {
                border: 1px solid #ddd;
                padding: 8px;
                word-wrap: break-word;
                max-width: 200px;
            }
            th {
                background-color: #f2f2f2;
            }
            td:nth-child(4) {
                width: 50%; /* 使abstract列的宽度是其他列的两倍 */
            }
        </style>
    </head>
    <body>
        <h1>Papers</h1>
        <table>
            <tr>
                <th>Title</th>
                <th>Authors</th>
                <th>Year</th>
                <th>Abstract</th>
                # <th>Keywords</th>
                <th>URL</th>
            </tr>
    ''')

    for paper in papers:
        htmlfile.write(f'''
            <tr>
                <td>{paper['title']}</td>
                <td>{', '.join(paper['authors'])}</td>
                <td>{paper['year']}</td>
                <td>{paper['abstract']}</td>
                <td><a href="{paper['url']}">{paper['url']}</a></td>
            </tr>
        ''')

    htmlfile.write('''
        </table>
    </body>
    </html>
    ''')

print(f"{len(papers)}信息已成功提取并保存到 {paper_dir}.html 文件中。")
```
